```python
import pandas as pd
```

```python
data1 = pd.read_csv('data/ECB_Notice_of_Violations__Cooling_towers_.csv', 
                    converters={'Violation Location (Block No.)': lambda x: str(x)})
data1.head()
```

```python
data1['Violation Location (Lot No.)'].fillna(0, inplace=True)
```

```python
data1['Violation Location (Lot No.)']=data1['Violation Location (Lot No.)'].astype(int).astype(str)
```

```python
data1['Violation Location (Lot No.)']=data1['Violation Location (Lot No.)'].str.zfill(4)
```

```python
data1.head()
```

```python
data1['Violation Location (Borough)'].unique()
```

```python
mapping = {'MANHATTAN':1,'BRONX':2,'BROOKLYN':3,'QUEENS':4,'STATEN IS':5}
data2 = data1.replace({'Violation Location (Borough)': mapping})
```

```python
data2.head()
```

```python
data2['BBL'] = data2['Violation Location (Borough)'].astype(str).str.cat(data2['Violation Location (Block No.)']).str.cat(data2['Violation Location (Lot No.)'])
```

```python
data2.head()
```

```python
data2.to_csv('data/Violation_CoolingTowers_BBL.csv')
```

```python
# Filter the violations to retreive only failure to register

dfCoolingTowersRegistrationViolations = \
data2[data2['Violation Details'].str.contains("FAILURE TO REGISTER")==True]
```
